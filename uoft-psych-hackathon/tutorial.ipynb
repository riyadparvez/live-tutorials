```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
df = pd.DataFrame(np.c_[dataset['data'], dataset['target']], columns= dataset['feature_names'] + ['target'])
df.head()
```

```python
df.columns[df.isnull().any()]
```

```python
df.describe()
```

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
df[features].head()
```

```python
df['dsfsdf'] = 1
df.head()
```

```python
df = df.drop(columns=['dsfsdf'])
df.head()
```

```python
df.groupby('target').agg(['mean', 'max', 'min'])
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import MinMaxScaler

mapper = DataFrameMapper(
    [
        (['sepal length (cm)'], MinMaxScaler(), {'alias': 'sepal length (cm)_scaled'}),
        (['sepal width (cm)'], MinMaxScaler(), {'alias': 'sepal width (cm)_scaled'}),
        (['petal length (cm)'], MinMaxScaler(), {'alias': 'petal length (cm)_scaled'}),
        (['petal width (cm)'], MinMaxScaler(), {'alias': 'petal width (cm)_scaled'}),
        (['target'], None),
    ],
    df_out=True
)

df_train =  mapper.fit_transform(df)
df_train
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_train, test_size=0.2)
train.head()
```

```python
from sklearn.linear_model import LogisticRegression

features = ['sepal length (cm)_scaled', 'sepal width (cm)_scaled', 'petal length (cm)_scaled', 'petal width (cm)_scaled']


log_reg = LogisticRegression()

# fit -> train
# features -> X, targets -> y
# predictions -> y_hat

log_reg = log_reg.fit(train[features], train['target'])
```

```python
predictions = log_reg.predict(test[features])
predictions
```

```python
from sklearn.metrics import classification_report

print(classification_report(test['target'], predictions))
```
